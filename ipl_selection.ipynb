```python
 #Split the 'BBI' column into two new columns
bowl_df[['BBI_wkts', 'BBI_bowls']] = bowl_df['BBI'].str.split('/', expand=True)

# Convert the new columns to numeric values
bowl_df['BBI_wkts'] = pd.to_numeric(bowl_df['BBI_wkts'])
bowl_df['BBI_bowls'] = pd.to_numeric(bowl_df['BBI_bowls'])

# For cases where the first value is greater than the second value, swap the columns
condition = bowl_df['BBI_wkts'] > bowl_df['BBI_bowls']
bowl_df.loc[condition, ['BBI_wkts', 'BBI_bowls']] = bowl_df.loc[condition, ['BBI_bowls', 'BBI_wkts']]

# Drop the original 'BBI' column
bowl_df.drop('BBI', axis=1, inplace=True)

bowl_df
```

```python
# Identify rows where 'low' > 'high'
swap_condition = bowl_df['BBI_wkts'] > bowl_df['BBI_bowls']

# Create temporary columns to store the swapped values
bowl_df['temp_low'] = bowl_df['BBI_wkts']
bowl_df['temp_high'] = bowl_df['BBI_bowls']

# Swap values where condition is met
bowl_df.loc[swap_condition, 'BBI_wkts'] = bowl_df.loc[swap_condition, 'temp_high']
bowl_df.loc[swap_condition, 'BBI_bowls'] = bowl_df.loc[swap_condition, 'temp_low']

# Drop temporary columns
bowl_df.drop(['temp_low', 'temp_high'], axis=1, inplace=True)

bowl_df
```

```python
# POS	Player	Mat	Inns	Ov	Runs	Wkts	BBI	Avg	Econ	SR	4w	5w
# List of columns to include in the aggregation
include_columns = ['Player', 'Mat', 'Inns', 'Ov', 'Runs', 'Wkts', '4w', '5w']

# Group by 'Player' and sum the specified columns for each player
final_bowls = bowl_df.groupby('Player')[include_columns].sum().reset_index()

# Print the resulting DataFrame
print(final_bowls)
```

```python
#adding total bowls column of the player
final_bowls['Total_bowls'] = final_bowls['Ov'] * 6
```

```python
# Calculate bowling Average
final_bowls['Avg'] = final_bowls['Runs'] / final_bowls['Wkts']

# Assuming 'wkts' column is already calculated
#When ascending is set to False, it means that the ranking will be done in descending order. 
#In other words, the higher values will be ranked higher.
final_bowls['Position'] = final_bowls['Wkts'].rank(ascending=False)

# Assuming 'total_bowls' and 'wkts' columns are available
final_bowls['SR'] = final_bowls['Total_bowls'] / final_bowls['Wkts']

final_bowls['Econ'] = final_bowls['Runs'] / (final_bowls['Total_bowls']/6)


final_bowls
```

```python
d1= bowl_df.groupby('Player')['BBI_wkts'].max().reset_index()
final_bowls['BBI_wkts'] = d1['BBI_wkts']
d2 = bowl_df.groupby('Player')['BBI_bowls'].min().reset_index()
final_bowls['BBI_bowls'] = d2['BBI_bowls']
```

```python
#sorting all the rows based on the position of the player which cal on the basis of the total wkts.
final_bowls = final_bowls.sort_values(by='Position', ascending=True)
# Reset the index of the DataFrame
final_bowls.reset_index(drop=True, inplace=True)

# Print the modified DataFrame
final_bowls
```
